```python
from __future__ import print_function
import os
import re
import sys
try:
    reload(sys)
    sys.setdefaultencoding('utf-8')
except:
    pass
```

```python
# 定义删除除字母,数字，汉字以外的所有符号的函数
def remove_punctuation(line):
    line = str(line)
    if line.strip() == '':
        return ''
    rule = re.compile(u"[^a-zA-Z0-9\u4E00-\u9FA5]")
    line = rule.sub('', line)
    return line
```

类别标签0-6分别对应：flood hail waterlogging gale drought thunderbolt typhoon

用最简单暴力的方式，依次使用相同的代码提取每个语料中的首行存储到data目录中

```python
# 提取洪水新闻标题 

target_file=open('./data/src_data/flood.txt','w',encoding='utf-8')                                          #汇总结果文件 
#提取data文件夹中所有文件的第一行
datapath = './语料/暴雨洪涝/'   
dirs = os.listdir(datapath)           #dirs得到所有txt文件名
for dir in dirs:
    fname = datapath+'\\'+dir
    with open(fname,'r',encoding='utf-8')as src_file:
        lines =src_file.readlines()
        first_line = lines[0] # 取第一行即新闻首行
        print(first_line) 
        handle_sentence = remove_punctuation(first_line)
        target_file.write(handle_sentence+'\t'+'0'+'\n')                                          #类别标签
target_file.close()
```
